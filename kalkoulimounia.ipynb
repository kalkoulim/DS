```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a count plot for the 'Diagnosis' target variable
plt.figure(figsize=(6, 4))
sns.countplot(x='Diagnosis', data=y, palette='viridis')
plt.title('Distribution of Diagnosis (Benign vs. Malignant)')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select a few key features for histograms
selected_features = ['radius1', 'texture1', 'perimeter1', 'area1', 'smoothness1']

# Create subplots for histograms
plt.figure(figsize=(15, 10))
for i, feature in enumerate(selected_features, 1):
    plt.subplot(2, 3, i) # Arrange plots in 2 rows, 3 columns
    sns.histplot(X[feature], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```
